```python
from kafka import KafkaConsumer
from time import sleep
from json import dumps,loads
import json
import boto3
```

```python
consumer = KafkaConsumer(
    'demotest',
     bootstrap_servers=['54.144.131.204:9092'],
    value_deserializer=lambda x: loads(x.decode('utf-8')))
```

```python
s3 = boto3.resource('s3')
```

```python
for bucket in s3.buckets.all():
    print(bucket.name)
```

```python
# Initialize the S3 bucket
s3_bucket_name = "de6007-group4-stockprediction"  # Modify the bucket name as needed
s3_bucket = s3.Bucket(s3_bucket_name)

# Iterate over Kafka messages
for count, message in enumerate(consumer):
    # Serialize message.value to JSON
    message_value_json = json.dumps(message.value)
    
    # Write JSON to S3 object
    s3_object_key = f"stock_market_{count}.json"  # Modify the file name as needed
    
    # Put the JSON data into the S3 bucket
    s3_bucket.put_object(
        Body=message_value_json,
        Key=s3_object_key
    )
```
